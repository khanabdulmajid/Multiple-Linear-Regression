```python
import pandas as pd

import numpy as np
```

```python
file=pd.read_csv("MultipleRegression.csv")  # reading data
```

```python
file.head(1)    # displaying 1 column
```

```python
w0=0.03
w1=0.04               # choosing random values for w0,w1,w2 for the equation yp= w0 + (w1 * x1) + (w2 * x2) 
w2=0.04
```

```python
x1=np.array(file["Interest Rate"])                      # value of x1
x2=np.array(file["Unemployment Rate"])                  # value of x2
```

```python
x1,x2  # diplaying x1 and x2
```

```python
y=np.array(file["Stock Price"])               # assigning a value to y
yp=w0 + w1*x1 + w2*x2                         # equation of ypredicted  
y,yp                                          # diplaying y,ypredicted
```

```python
n=len(yp)                   
j=(sum((y-yp)**2))/(2*n)                       # error calculation 
j
```

```python


error1=j
alpha=0.05
while (error1>1586):
    dj0=(sum(y-w0-w1*x1-w2*x2))/(n)
    dj1=(sum((y-w0-w1*x1-w2*x2)* x1))/(n)
    dj2=(sum((y-w0-w1*x1-w2*x2)*x2))/(n)
    
    w0=w0+alpha*dj0
    w1=w1+alpha*dj1
    w2=w2+alpha*dj2
    
    yp = w0 + w1*x1 + w2*x2
    
    error1=(sum((y-w0-w1*x1-w2*x2)**2))/(2*n)
    print('error ',error1,' w0 ',w0,' w1 ',w1,' w2 ',w2)
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file["Ypredicted"]=yp                # writing to the dataset
file
```

```python
# plotting Graphs with original values and predicted values

sns.lineplot(x1,y,label=('X1'))
sns.lineplot(x2,y,label=('X2'))
sns.lineplot(x1,yp,label=('X1Pred'))
sns.lineplot(x2,yp,label=('X2Pred'))
```
